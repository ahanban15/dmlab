```python
import seaborn as sns
 
sns.countplot(x = "Education", hue = "Loan_Status", data = loan_data)
plt.show() 
```

```python
df = loan_data.select_dtypes(include=['float64', 'int64'])
crrmtx = df.corr()

sns.heatmap(crrmtx, annot=True, fmt='.2f')
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Generate some random data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Create a DataFrame
df = pd.DataFrame(data, columns=['CoapplicantIncome'])

# Create Q-Q plot
sm.qqplot(df['CoapplicantIncome'], line ='45')
plt.title('Q-Q Plot')
plt.show()
```
