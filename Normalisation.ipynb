```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'loan_train.csv'
loan_data = pd.read_csv(file_path)


loan_data.head(10)
```

```python
import seaborn as sns
sns.histplot(loan_data['ApplicantIncome'])
```

```python
import numpy as np
mean_appinc = loan_data['ApplicantIncome'].mean()

stddev_appinc = loan_data['ApplicantIncome'].std()

loan_data['norm_zscore'] = (loan_data['ApplicantIncome'] - mean_appinc) / stddev_appinc

print("Mean of ApplicantIncome is: ", mean_appinc)
print("Standard Deviation of ApplicantIncome is: ", stddev_appinc)

print("Normalised data:")
loan_data['norm_zscore']

loan_data.head(10)
```

```python
#sns.barplot(loan_data['norm_zscore'])
sns.histplot(loan_data['norm_zscore'])
```

```python
min_appinc = loan_data['ApplicantIncome'].min()

max_appinc = loan_data['ApplicantIncome'].max()

loan_data['norm_minmax'] = (loan_data['ApplicantIncome'] - min_appinc) / (max_appinc - min_appinc)
print("Min of ApplicantIncome is: ", min_appinc)
print("Max of ApplicantIncome is: ", max_appinc)

print("Normalised data:")
loan_data['norm_minmax']

loan_data.head(10)
```

```python
#sns.barplot(loan_data['norm_minmax'])
sns.histplot(loan_data['norm_minmax'])
```

```python
loan_data['norm_meannorm'] = (loan_data['ApplicantIncome'] - mean_appinc) / (max_appinc - min_appinc)
print("Min of ApplicantIncome is: ", min_appinc)
print("Max of ApplicantIncome is: ", max_appinc)
print("Mean of ApplicantIncome is: ", mean_appinc)

print("Normalised data:")
loan_data['norm_meannorm']

loan_data.head(10)
```

```python
#sns.barplot(loan_data['norm_meannorm'])
sns.histplot(loan_data['norm_meannorm'])
```

```python
loan_data['norm_maxabs'] = (loan_data['ApplicantIncome'] / abs(max_appinc))

print("Normalised data:")
loan_data['norm_maxabs']

loan_data.head(10)
```

```python
#sns.barplot(loan_data['norm_maxabs'])
sns.histplot(loan_data['norm_maxabs'])
```

```python
import numpy as np
from scipy import stats
median_appinc = loan_data['ApplicantIncome'].median()

iqr_appinc = stats.iqr(loan_data['ApplicantIncome'], interpolation = 'midpoint')

loan_data['norm_robust'] = (loan_data['ApplicantIncome'] - median_appinc)/ iqr_appinc

print("Normalised data:")
loan_data['norm_robust']

loan_data.head(10)
```

```python
#sns.barplot(loan_data['norm_robust'])
sns.histplot(loan_data['norm_robust'])
```
