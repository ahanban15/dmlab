```python
from math import *
from decimal import Decimal
from numpy import random
import matplotlib.pyplot as plt
import numpy as np
from math import degrees, acos
```

```python
import matplotlib.pyplot as plt
def euclidean_distance(x, y):
    return sqrt(sum(pow(a - b, 2) for a, b in zip(x, y)))

x = random.randint(100, size=2)
y = random.randint(100, size=2)

print("x = ", x)
print("y = ", y)
print("Euclidean distance: ", euclidean_distance(x, y))
plt.scatter(x[0], x[1], color='blue', label='Point x')
plt.scatter(y[0], y[1], color='red', label='Point y')
plt.plot([x[0], y[0]], [x[1], y[1]], linestyle='--', color='green', label='Euclidean Distance')
plt.legend()
plt.title('Euclidean Distance between Points x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Manhattan Distance
def manhattan_distance(x, y):
    return [abs(a - b) for a, b in zip(x, y)]

x = random.randint(100, size=2)
y = random.randint(100, size=2)

print("x = ", x)
print("y = ", y)
manhattan_dist = manhattan_distance(x, y)
print("Manhattan distance : ",sum(manhattan_dist))

dimensions = ['X', 'Y']
# Plot the points
plt.plot([x[0], x[0]], [x[1], x[1]], 'ro--', label='Horizontal Distance')
plt.plot([y[0], y[0]], [y[1], y[1]], 'bo--', label='Vertical Distance')

# Plot the horizontal distance
plt.plot([x[0], y[0]], [x[1], x[1]], 'r--')
plt.text((x[0] + y[0]) / 2, x[1] - 0.1, str(abs(x[0] - y[0])), ha='center')

# Plot the vertical distance
plt.plot([y[0], y[0]], [x[1], y[1]], 'r--')
plt.text(y[0] + 0.1, (x[1] + y[1]) / 2, str(abs(x[1] - y[1])), va='center')

plt.title('Manhattan Distance between x and y')
plt.xlabel('Dimension')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```

```python
# Minkowski Distance
def nth_root(value, n_root):
    return round((value ** (1/n_root)), 3)

def minkowski_distance(x, y, p_value):
    return nth_root(sum(pow(abs(a - b), p_value) for a, b in zip(x, y)), p_value)

print("x = ", x)
print("y = ", y)

p_value = 2
minkowski_dist = minkowski_distance(x, y, p_value)
print("Minkowski distance for p = {}: {}".format(p_value, minkowski_dist))
plt.scatter(x[0], x[1], color='blue', label='Point x')
plt.scatter(y[0], y[1], color='red', label='Point y')
plt.plot([x[0], y[0]], [x[1], y[1]], linestyle='--', color='green', label='Minkowski Distance (p={})'.format(p_value))
plt.legend()
plt.title('Minkowski Distance (p={}) between Points x and y'.format(p_value))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#Cosine Similarity
def square_rooted(x):

    return round(sqrt(sum([a*a for a in x])),3)

def cosine_similarity(x,y):

   numerator = sum(a*b for a,b in zip(x,y))
   denominator = square_rooted(x)*square_rooted(y)
   return round(numerator/float(denominator),3)

print("x = ",x)
print("y = ",y)
print("Cosine Similarity : ",cosine_similarity(x,y))
cos_sim = np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))
angle_degrees = degrees(acos(cos_sim))

plt.figure(figsize=(8, 6))
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector x')
plt.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector y')

plt.plot([0, y[0]], [0, y[1]], linestyle='--', color='gray')

plt.annotate(f'Angle: {angle_degrees:.2f} degrees', xy=(y[0]/2, y[1]/2), xytext=(10, 10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->", color='black'))

plt.title('Vectors and Angle Between Them')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
