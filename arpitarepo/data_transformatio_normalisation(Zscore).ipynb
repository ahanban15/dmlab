Min-Max Normalisation

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')

# Extract 'Age' column
age = df['Age']

# Calculate min and max for 'Age'
min_age, max_age = age.min(), age.max()

# Normalize 'Age' using the Min-Max scaling formula
normalized_age = (age - min_age) / (max_age - min_age)

# Plot histogram for normalized 'Age'
plt.hist(normalized_age, bins=20, alpha=0.5, color='purple', edgecolor = 'black')
plt.title('Histogram of Min-Max Normalized Age')
plt.xlabel('Normalized Age')
plt.ylabel('Frequency')
plt.show()
```

Mean Normalised

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')

# Extract 'Age' column
age = df['Age']

# Calculate mean and range for 'Age'
mean_age, range_age = age.mean(), age.max() - age.min()

# Mean normalize 'Age'
normalized_age = (age - mean_age) / range_age

# Plot histogram for mean-normalized 'Age'
plt.hist(normalized_age, bins=20, alpha=0.5, color='blue', edgecolor = 'black')
plt.title('Histogram of Mean-Normalized Age')
plt.xlabel('Mean-Normalized Age')
plt.ylabel('Frequency')
plt.show()
```

Robust Scale

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('train.csv')

# Extract 'Age' column and remove NaN values
age = df['Age'].dropna()

original_mean_age = age.mean()
original_std_dev_age = age.std()

median_age = age.median()
iqr_age = age.quantile(0.75) - age.quantile(0.25)     # Calculate IQR for 'Age'

scaled_age = (age - median_age) / iqr_age
scaled_std_age = scaled_age.std()

# Create subplots for original and normalized histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot histogram for original 'Age'
ax1.hist(age, bins=20, alpha=0.5, color='blue', edgecolor = 'black')
ax1.set_title('Histogram of Original Age')
ax1.set_xlabel('Original Age')
ax1.set_ylabel('Frequency')

# Plot histogram for normalized 'Age'
ax2.hist(scaled_age, bins=20, alpha=0.5, color='brown', edgecolor = 'black')
ax2.set_title('Histogram of Robust Scaled Age')
ax2.set_xlabel('Robust Scaled Age')
ax2.set_ylabel('Frequency')
plt.show()

# Print original and normalized mean
print("\nStatistics:")
print(f"Original Mean Age: {original_mean_age:.2f}")

# Print normalized data
print("\Original Data:")
print(age)

# Print normalized data
print("\nNormalized Data:")
print(scaled_age)

# Print original and scaled standard deviations
print(f"\nOriginal Standard Deviation of Age: {original_std_age:.2f}")
print(f"Scaled Standard Deviation of Robust Scaled Age: {scaled_std_age:.2f}")
```
