```python

'''

Minkowski distance is a more general distance metric that includes both Euclidean and Manhattan distances as special cases. It considers the absolute differences between corresponding components of vectors. On the other hand, cosine similarity focuses on the angle between vectors and is not influenced by the magnitude of the vectors, making it effective in scenarios where the magnitude is less relevant, such as text analysis.

Cosine similarity

Cosine similarity is a measure of similarity, often used to measure document similarity in text analysis. We use the below formula to compute the cosine similarity.

Similarity = (A.B) / (||A||.||B||)

where A and B are vectors:

    A.B is dot product of A and B: It is computed as sum of element-wise product of A and B.

    ||A|| is L2 norm of A: It is computed as square root of the sum of squares of elements of the vector A.


'''
```

```python
import numpy as np
from scipy.spatial.distance import minkowski

point1 = np.array([2, 3])
point2 = np.array([5, 7])

# Euclidean Distance
euclidean_distance = np.linalg.norm(point1 - point2)

# Manhattan Distance
manhattan_distance = np.sum(np.abs(point1 - point2))

# Minkowski Distance (setting p=2 for Euclidean distance)
minkowski_distance = minkowski(point1, point2, p=2)

# Cosine Similarity
cosine_similarity = np.dot(point1, point2) / (np.linalg.norm(point1) * np.linalg.norm(point2))

# Displaying the results
print(f"Euclidean Distance: {euclidean_distance}")
print(f"Manhattan Distance: {manhattan_distance}")
print(f"Minkowski Distance (p=2): {minkowski_distance}")
print(f"Cosine Similarity: {cosine_similarity}")
```

```python
# Euclidean Distance

import numpy as np
import matplotlib.pyplot as plt

point1 = np.random.randint(0,100, size = 2)
point2 = np.random.randint(0,100, size = 2)

print("X = ", point1)
print("Y = ", point2)

#Calculating euclidean distance
sum_sq = np.sum(np.square(point1-point2))
euc_distance = np.sqrt(sum_sq)

print(" Euclidean distance: ",  euc_distance )
plt.plot(point1, point2, marker = 'o')


plt.title("Euclidean Distance")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
